### Question 5

How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?


**Options:**
 - 48
 - 38
 - 14
 - 1

```python
%%sql
select count(distinct m.movie_id)  from movies m
inner join genremap gm
on m.movie_id = gm.movie_id
inner join genres g
on g.genre_id = gm.genre_id
left join keywordmap km
on km.movie_id = m.movie_id
left join keywords k
on k.keyword_id = km.keyword_id
where genre_name = "Thriller" and keyword_name like '%love%'
```

### Question 7

### Question 9

What are the three production companies that have the highest movie popularity score on average, as recorded within the database?


**Options:**

 - MCL Films S.A., Turner Pictures, and George Stevens Productions
 - The Donners' Company, Bulletproof Cupid, and Kinberg Genre
 - Bulletproof Cupid, The Donners' Company, and MCL Films S.A
 - B.Sting Entertainment, Illumination Pictures, and Aztec Musique

```python
%%sql
select production_company_name, avg(popularity) as pop_index
from productioncompanies p
inner join productioncompanymap pm
on p.production_company_id = pm.production_company_id
inner join movies m
on m.movie_id = pm.movie_id
group by production_company_name
order by pop_index desc
limit 3
```

### Question 11

Which genre has, on average, the lowest movie popularity score? 


**Options:**

 - Science Fiction
 - Animation
 - Documentary
 - Foreign

```python
%%sql
select genre_name, avg(popularity) as pop_index
from genres g
inner join genremap gm
on g.genre_id = gm.genre_id
inner join movies m
on m.movie_id = gm.movie_id
group by genre_name
order by pop_index
limit 1
```
